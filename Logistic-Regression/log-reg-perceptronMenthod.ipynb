PERCEPTRON METHOD - A Perceptron is the simplest neural network model introduced by Frank Rosenblatt in 1958.

It is a binary classifier, meaning it predicts one of two classes:

0 or 1
Yes or No
Spam or Not Spam

It is the foundation of modern neural networks.

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
        
def step(z):
    return 1 if z>0 else 0
intercept_,coef_ = perceptron(X,y)
print("coefficient:", coef_)
print("intercept:", intercept_)
```

Both models compute a weighted sum of the input features. The Perceptron applies a step activation function and directly predicts class 0 or 1. Logistic Regression replaces the step function with the Sigmoid function, producing a probability between 0 and 1. This allows Logistic Regression to express prediction confidence and be trained using Log Loss and Gradient Descent, making it more flexible and suitable for many real-world classification tasks.
