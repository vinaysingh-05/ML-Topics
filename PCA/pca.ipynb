```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bike-sharing.csv")[["mnth","hr","weekday","temp","atemp"]]
```

```python
df
```

```python
import plotly.express as px

fig = px.scatter_3d(df, x="temp", y="atemp", z="hr")

fig.update_traces(marker=dict(
    size=5,
    color=df["atemp"],   # ✅ numeric (no conversion)
    colorscale='Viridis',
    showscale=True
))

fig.show()
```
