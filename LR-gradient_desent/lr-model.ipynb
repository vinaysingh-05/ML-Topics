```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train.reshape(-1, 1), y_train)
```

```python
m = lr.coef_
c = lr.intercept_
```

```python
print ("m =", lr.coef_)
print("c =", lr.intercept_)
```

```python
#by gradient descent

class GDRegressor : 
    def __init__(self, learning_rate=0.01 , epochs=100):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.m = m
        self.c = c

    def fit(self,x,y):
        n = len(x)
        for _ in range(self.epochs):
            loss_slope_C = -2*np.sum(y-self.m*x.ravel()-self.c)
            loss_slope_M = -2*np.sum((y-self.m*x.ravel()-self.c)*x.ravel())
            self.c = self.c - (self.learning_rate * loss_slope_C)
            self.m = self.m - (self.learning_rate * loss_slope_M)

            print(self.m , self.c)

            def predict(self,x):
                return self.m*x + self.c
```

```python
gd = GDRegressor(learning_rate=0.001, epochs=100)
gd.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, lr.predict(x_test.reshape(-1, 1)))
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# -----------------------------
# Sklearn Linear Regression
# -----------------------------
y_pred_lr = lr.predict(x_test.reshape(-1,1))

print("===== Sklearn Linear Regression =====")
print("R2 Score :", r2_score(y_test, y_pred_lr))
print("MAE      :", mean_absolute_error(y_test, y_pred_lr))
print("MSE      :", mean_squared_error(y_test, y_pred_lr))
print("RMSE     :", np.sqrt(mean_squared_error(y_test, y_pred_lr)))

# -----------------------------
# Gradient Descent Model
# -----------------------------
y_pred_gd = gd.m.item() * x_test + gd.c

print("\n===== Gradient Descent =====")
print("R2 Score :", r2_score(y_test, y_pred_gd))
print("MAE      :", mean_absolute_error(y_test, y_pred_gd))
print("MSE      :", mean_squared_error(y_test, y_pred_gd))
print("RMSE     :", np.sqrt(mean_squared_error(y_test, y_pred_gd)))
```

```python
plt.figure(figsize=(8,5))

plt.scatter(x_test, y_test, color='blue', label='Actual Data')

plt.plot(x_test, y_pred_lr, color='green', linewidth=2, label='Linear Regression')

plt.plot(x_test, y_pred_gd, color='red', linewidth=2, label='Gradient Descent')

plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Model Comparison")

plt.legend()
plt.show()
```
