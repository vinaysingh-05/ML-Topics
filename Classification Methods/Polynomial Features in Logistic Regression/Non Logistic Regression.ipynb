This is one of the most important concepts in Logistic Regression.

Many beginners think:

"Logistic Regression can only solve linear problems."

That's not completely true.

If we add Polynomial Features, Logistic Regression can also learn non-linear decision boundaries.

When Should You Use It?

Use Polynomial Features with Logistic Regression when:

The data is not linearly separable.
A normal Logistic Regression model performs poorly because the decision boundary needs to be curved.
You want to improve a simple linear model before moving to more complex algorithms.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:,2].values
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()

```

```python
# use the categorical species column as target (not petal length)
y = df['species'].values
clf.fit(X, y)
```

```python
from sklearn.model_selection import cross_val_score
y = np.asarray(y)
np.mean(cross_val_score(clf, X, y, scoring='accuracy', cv=10))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3,include_bias=False)
X_trf = poly.fit_transform(X)
```

```python
clf1 = LogisticRegression()
np.mean(cross_val_score(clf1,X_trf,y,scoring='accuracy',cv=10))
```

```python
def plot_decision_boundary(X,y,degree=1):
    
    poly = PolynomialFeatures(degree=degree)
    X_trf = poly.fit_transform(X)
    
    clf = LogisticRegression()
    clf.fit(X_trf,y)
    
    accuracy = np.mean(cross_val_score(clf,X_trf,y,scoring='accuracy',cv=10))
    
    a=np.arange(start=X[:,0].min()-1, stop=X[:,0].max()+1, step=0.01)
    b=np.arange(start=X[:,1].min()-1, stop=X[:,1].max()+1, step=0.01)


    XX,YY=np.meshgrid(a,b)
    
    input_array=np.array([XX.ravel(),YY.ravel()]).T

    labels=clf.predict(poly.transform(input_array))
    
    plt.contourf(XX,YY,labels.reshape(XX.shape),alpha=0.5)
    plt.scatter(X[:,0],X[:,1], c=y)
    plt.title('Degree = {}, accuracy is {}'.format(degree,np.round(accuracy,4)))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_enc = le.fit_transform(y)

plot_decision_boundary(X[:, :2], y_enc, degree=2)
```
