# Logistic Regression Solvers Comparison (Scikit-learn)

| **Penalties / Behaviors** | **liblinear** | **lbfgs** | **newton-cg** | **sag** | **saga** |
|---------------------------|:-------------:|:---------:|:-------------:|:-------:|:--------:|
| **Multinomial + L2 penalty** | No | Yes | Yes | Yes | Yes |
| **OVR + L2 penalty** | Yes | Yes | Yes | Yes | Yes |
| **Multinomial + L1 penalty** | No | No | No | No | Yes |
| **OVR + L1 penalty** | Yes | No | No | No | Yes |
| **Elastic-Net** | No | No | No | No | Yes |
| **No penalty (`none`)** | No | Yes | Yes | Yes | Yes |
| **Penalize the intercept (bad)** | Yes | No | No | No | No |
| **Faster for large datasets** | No | No | No | Yes | Yes |
| **Robust to unscaled datasets** | Yes | Yes | Yes | No | No |

## Key Takeaways

- **`liblinear`**
  - Supports **L1** and **L2** penalties.
  - Best for **small datasets**.
  - Does **not** support multinomial logistic regression.
  - Works well even if features are **not scaled**.

- **`lbfgs`**
  - Default solver in Scikit-learn.
  - Supports **L2** and **no penalty**.
  - Supports **multinomial** classification.
  - Requires **feature scaling** for best performance.

- **`newton-cg`**
  - Similar to `lbfgs`.
  - Supports **L2** and **no penalty**.
  - Good for **multinomial** problems.

- **`sag`**
  - Optimized for **large datasets**.
  - Supports **L2** and **no penalty**.
  - Requires **scaled features**.

- **`saga`**
  - Most versatile solver.
  - Supports **L1**, **L2**, **Elastic-Net**, and **no penalty**.
  - Supports **multinomial** logistic regression.
  - Best choice for **large datasets** with sparse or high-dimensional features.
