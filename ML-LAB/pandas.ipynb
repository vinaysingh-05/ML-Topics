EXP-03 : Implementation of Data Preprocessing and DataFrame Concept using Pandas for Machine Learning

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [20, 21, np.nan, 23, 24],
    'Salary': [20000, 25000, 30000, np.nan, 40000],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR']
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
# Display first 5 rows
print("\nFirst 5 Rows:\n", df.head())

# Display dataset information (data types, non-null values)
print("\nDataset Info:")
print(df.info())

# Statistical summary
print("\nStatistical Summary:\n", df.describe())
```

```python
#CHECK MISSING VALUES 

# Check for null values
print("\nMissing Values:\n", df.isnull())

# Count missing values
print("\nMissing Values Count:\n", df.isnull().sum())
```

```python
# HANDLE MISSING VALUES 

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)

print("\nAfter Handling Missing Values:\n", df)
```

```python
# REMOVE DUPLICATES 
df = df.drop_duplicates()
# Convert Salary column to integer type
df = df.drop_duplicates()
df['Salary'] = df['Salary'].fillna(df['Salary'].median()).astype(int)

# Select rows where Age > 21
filtered_data = df[df['Age'] > 21]
print("\nFiltered Data (Age > 21):\n", filtered_data)
```

```python
# Add new column (Bonus = 10% of Salary)
df['Bonus'] = df['Salary'] * 0.10

print("\nDataFrame with Bonus Column:\n", df)
```

```python
print("\nFinal Processed DataFrame:\n", df)
```

EXP-04 Implement data preprocessing (for handle missing values by deletion and imputation, for handle missing values by LabelEncoder & OneHotEncoder) using scikitlearn.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
data = {
    'Name': ['A', 'B', 'C', 'D'],
    'Age': [20, np.nan, 22, 23],
    'Salary': [20000, 25000, np.nan, 30000],
    'Department': ['HR', 'IT', 'HR', 'Finance']
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

df_deleted = df.dropna()
print("\nAfter Deletion:\n", df_deleted)

# IMPUTATION 
imputer = SimpleImputer(strategy='mean')
df[['Age', 'Salary']] = imputer.fit_transform(df[['Age', 'Salary']])
print("\nAfter Imputation:\n", df)

# LABEL ENCODING 

le = LabelEncoder()
df['Department_Label'] = le.fit_transform(df['Department'])

print("\nAfter Label Encoding:\n", df)

# ONE HOT ENCODING 

one_hot = pd.get_dummies(df['Department'])

print("\nOne Hot Encoded Data:\n", one_hot)

# Combine with original DataFrame
df = pd.concat([df, one_hot], axis=1)

print("\nFinal DataFrame:\n", df)
```
