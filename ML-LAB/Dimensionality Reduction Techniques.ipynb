EXP-07  Implementation of Dimensionality Reduction Techniques (PCA, LDA) and Hyperparameter Tuning (GridSearchCV) using Scikit-learn

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
```

```python
data = load_iris()
X = data.data
y = data.target

# ---------------- SPLIT DATA ----------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# ---------------- PCA ----------------
# Reduce features to 2 components
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

print("PCA Transformed Data:\n", X_pca[:5])
```

```python
# ---------------- LDA ----------------
# Reduce to 2 components using class labels
lda = LinearDiscriminantAnalysis(n_components=2)

X_lda = lda.fit_transform(X, y)

print("\nLDA Transformed Data:\n", X_lda[:5])
```

```python
# ---------------- MODEL TRAINING ----------------
model = SVC()
```

```python
# ---------------- GRID SEARCH ----------------
# Define parameter grid
param_grid = {
    'C': [1, 10, 100],
    'kernel': ['linear', 'rbf']
}

# Apply GridSearchCV
grid = GridSearchCV(model, param_grid, cv=5)

# Train model
grid.fit(X_train, y_train)

# Best parameters
print("\nBest Parameters:", grid.best_params_)
```

```python
# Best model prediction
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)

print("Prediction using Best Model:", y_pred)
```
