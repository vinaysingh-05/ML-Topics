# 📖 Theory

## What is Decision Tree Regression?

Decision Tree Regression is a supervised machine learning algorithm used for predicting continuous values.

It recursively splits the dataset into smaller regions based on feature values to minimize prediction error.

Unlike Linear Regression, it can learn non-linear relationships.

### Advantages
- No Feature Scaling Required
- Handles Non-linear Data
- Easy to Interpret
- Works with Numerical & Categorical Data

### Disadvantages
- Can Overfit
- Sensitive to Noise
- High Variance

# 📂 Step 1 : Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
data = fetch_california_housing()

X = data.data

y = data.target
```

```python
df = pd.DataFrame(
    data.data,
    columns=data.feature_names
)

df["Price"] = data.target

df.head()
```

```python
df = pd.DataFrame(
    data.data,
    columns=data.feature_names
)

df["Price"] = data.target

df.head()
```

```python
plt.figure(figsize=(7,5))

plt.hist(df["Price"], bins=30)

plt.title("House Price Distribution")

plt.xlabel("Price")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    df["MedInc"],
    df["Price"]
)

plt.xlabel("Median Income")

plt.ylabel("House Price")

plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42

)
```

```python
# Create Decision Tree Model

model = DecisionTreeRegressor(random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
# Predict House Prices

y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

print("MAE :", mae)

print("MSE :", mse)

print("RMSE :", rmse)

print("R² :", r2)
```

```python
results = pd.DataFrame({

    "Actual":y_test,

    "Predicted":y_pred

})

results.head(10)
```
